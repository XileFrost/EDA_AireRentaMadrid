##### **INTRODUCCIÓN**

Se asocian los países más desarrollados e industrializados con los más contaminantes a nivel mundial. Partiendo de esta premisa, en este EDA se quiere hacer una hipotética extrapolación local, a la ciudad en la que vivimos, Madrid, y **se quiere demostrar que en los distritos con rentas más altas de Madrid ciudad la calidad del aire es peor**. Asimismo, se plantean cuatros hipótesis adicionales: 
- A mayor densidad de población, peor es la calidad del aire. 
- Los hábitos de movilidad de la población de un distrito pueden favorecer o perjudicar la calidad del aire. 
- A mayor extensión de zonas verdes en el distrito, mejor calidad del aire. 
- A mayor número de habitantes por vehículo en el distrito, mejor calidad del aire. 

Para confirmar o negar mis hipótesis voy a valorar diferentes aspectos relacionados con la calidad del aire a partir de series de datos de diversa índole: 

Datos con carácter geoposicional: 
- Estaciones de medición de la calidad del aire en Madrid ciudad en 2023. 

Datos con caracteres medioambiental y ecológico: 
- Datos anuales de la calidad del aire en Madrid ciudad en 2023. 
- Superficie de parques y zonas verdes por distrito en Madrid ciudad en 2023. 

Datos con carácter socioeconómico: 
- Renta bruta por persona y hogar de los distritos de Madrid ciudad 2022.

Datos con carácter demográfico: 
- Población en Madrid por distrito a 1 de enero de 2023. 

Datos con caracteres movilidad y transporte: 
- Padrón de vehículos en Madrid ciudad por distrito en 2024 a partir de Impuesto de vehículos de tracción mecánica 2023.  
- Medios de transporte utilizados habitualmente por distrito en 2023.

##### **DATASET 2 - Estaciones de medición de la Calidad del Aire en Madrid ciudad, año 2023**

```python
df_dataset2 = pd.read_csv('./data/02_datos-aire-anuales-2023.csv', sep=';')
df_dataset2
```

En tan sólo un vistazo veo que en los dos primeros dataframes sólo hay una columna coincidente, 'ESTACION - Estación', a partir de ella unificaré el contenido de ambos dataframes, para lo cual en la cabecera de las columnas del dataset 2 he de convertir los títulos a mayúsculas y suprimir tildes. 

```python
# Convierto a mayúsculas los títulos de las columnas y suprimo tildes.
df_dataset2.columns = df_dataset2.columns.str.upper()
df_dataset2.columns = df_dataset2.columns.str.replace('ESTACIÓN','ESTACION')
df_dataset2.columns = df_dataset2.columns.str.replace('CONCENTRACIONES MÁXIMAS HORARIAS', 'CONCENTRACIONES MAXIMAS HORARIAS')
df_dataset2
```

```python
# Unifico los dos dataframes en un nuevo dataframe a partir de la columna 'ESTACION'
df_aire = pd.merge(df_estaciones_info, df_dataset2, on='ESTACION', how='inner') 
df_aire
```

```python
# Reordeno las columnas, para que ZONA esté primero y ESTACION en segundo lugar.
cols = ['ZONA', 'ESTACION'] + [col for col in df_aire.columns if col not in ['ZONA', 'ESTACION']]
df_aire = df_aire[cols]
# Ordeno el dataframe por 'ZONA' y luego por 'ESTACION'.
df_aire = df_aire.sort_values(by=['ZONA', 'ESTACION'], ascending=[True, True])
df_aire = df_aire.reset_index(drop=True)
df_aire
```

El actual [protocolo de actuación del Ayuntamiento de Madrid para episodios de alta contaminación](https://www.madrid.es/portales/munimadrid/es/Inicio/Medidas-especiales-de-movilidad/Protocolo-de-contaminacion/Protocolo-de-actuacion-para-episodios-de-contaminacion-por-dioxido-de-nitrogeno/?vgnextfmt=default&vgnextoid=fd8718cea863c410VgnVCM1000000b205a0aRCRD&vgnextchannel=00b3cf7588c97610VgnVCM2000001f4a900aRCRD), vigente desde el 10 de diciembre de 2018, se activa en función de las concentraciones de 'Dióxido de nitrógeno' (NO2). Esto podría explicar que dicho dato sea el único valor común medido por las 24 Estaciones de Medición de Calidad del Aire de la ciudad Madrid, por ello haré un último filtrado en el dataframe, quedándome sólo con las filas en las que esté presente la 'MAGNITUD' 'Dióxido de nitrógeno'. 

```python
df_aire = df_aire[df_aire['MAGNITUD'] == 'Dióxido de nitrógeno']
df_aire = df_aire.reset_index(drop=True)
df_aire
```

```python
# Me quedo sólo con las columnas 'ESTACION', 'CONCENTRACIONES MAXIMAS HORARIAS' y 'CONCENTRACIONES MEDIAS ANUALES'.
df_no2 = df_aire[['ESTACION', 'CONCENTRACIONES MAXIMAS HORARIAS']]
df_no2
```

##### **DATASET 3 - Renta bruta por persona y hogar de los distritos de Madrid ciudad, año 2022**

```python
df_dataset3 = pd.read_csv('./data/03_renta_bruta_media_persona_y_hogar.csv', sep=';')
df_dataset3
```

```python
# Elimino las columnas 'Municipios' y 'Secciones'.
df_dataset3= df_dataset3.drop(['Municipios','Secciones'], axis = 1 )
df_dataset3
```

Esto es muy farragoso, por tanto, voy a reorganizar los datos en nuevo dataframe, una fila por distrito y las columnas 'DISTRITO', 'RENTA BRUTA MEDIA POR PERSONA' y 'RENTA BRUTA MEDIA POR HOGAR'.

```python
# Filtro las filas que contienen los valores de renta bruta media.
df_rentas= df_dataset3[df_dataset3['Indicadores de renta media y mediana'].isin(['Renta bruta media por persona', 'Renta bruta media por hogar'])]
# Creo una pivot table que se ajuste a mis necesidades.
df_rentas_pivot = df_rentas.pivot_table(index='Distritos',
                                columns='Indicadores de renta media y mediana',
                                values='Total',
                                aggfunc='first')
# Reseteo los nombres de las columnas.
df_rentas_pivot.columns.name = None
df_rentas_pivot.reset_index(inplace=True)
# Renombro las columnas.
df_rentas_pivot.rename(columns={
    'Distritos' : 'DISTRITO',
    'Renta bruta media por persona': 'RENTA BRUTA MEDIA POR PERSONA',
    'Renta bruta media por hogar': 'RENTA BRUTA MEDIA POR HOGAR'
}, inplace=True)

df_rentas_pivot = df_rentas_pivot[['DISTRITO', 'RENTA BRUTA MEDIA POR PERSONA', 'RENTA BRUTA MEDIA POR HOGAR']]
df_rentas_pivot = df_rentas_pivot.iloc[:, 1:]
df_rentas_pivot
```

**Consideraciones relevantes:**

Voy a asignar de manera totalmente consciente la Estación Plaza Elíptica al Distrito 12 Usera porque, aunque se encuentra en el Distrito 11 Carabanchel, se encuentra a una sola manzana de formar parte de Usera. ##### **GRÁFICO 10: Heatmap para Variables Objeto de Estudio** <br>

Un heatpmap es muy buena herramienta para confirmar posibles correlaciones entre todas las variables numéricas implicadas en un Análisis de Datos. Esta interdependencia se verifica atendiendo al valor obtenido, que nos permitirá hablar de los siguientes tipos de correlaciones: 
- Correlaciones cercanas a 1: relación positiva fuerte. El incremento de una variable repercute en el aumento de otra o, en sentido contrario, la disminución de una variable implica el descenso de otra.   
- Correlaciones cercanas a -1: relación negativa fuerte. El incremento de una variable conlleva la bajada de otra y viceversa, la disminución de una variable trae consigo la subida de otra.  
- Correlaciones cercanas a 0: poca o ninguna relación. 

```python
# Cambio el nombre de la columna 'Densidad por KM2' para mejorar la interpretación de datos fuera de contexto
df_datosobjetoestudio.columns = df_datosobjetoestudio.columns.str.replace('DENSIDAD POR KM2', 'DENSIDAD DE POBLACIÓN')

# Selecciono sólo las columnas numéricas
df_estudionumerico = df_datosobjetoestudio.select_dtypes(include='number')

# Calculo la matriz de correlación
correlation_matrix = df_estudionumerico.corr()

# Configuro el gráfico
plt.figure(figsize=(14, 8))
sns.set_theme(style="whitegrid")

# Creo el heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, center=0, linewidths=1)
plt.title("")
plt.tight_layout()

plt.show()
```

**Correlaciones Positivas** (cercanas a 1): Número de Personas & Total Vehículo, Renta Bruta Media por Persona & Renta Bruta Media por Hogar. <br>

**Correlaciones Negativas** (cercanas a -1): Densidad de Población & Uso Coche Particular, Densidad de Población & hectáreas de Zonas Verdes y Parques, Renta Bruta Media por Persona & Hab/Coche, Renta Bruta Media por Hogar & Hab/Coche <br>

**Correlaciones Cercanas a 0**: Habitantes & Densidad de Población, Habitantes & NO2, Densidad de Población & Renta Bruta Media por Hogar, Densidad de Población & Total Vehículos, Densidad de Población & Hab/Coche, Densidad de Población & NO2, Renta Bruta Media por Hogar & Uso Coche Particular, Hab/Coche & Uso Coche Particular, Uso Coche Particular & NO2.
